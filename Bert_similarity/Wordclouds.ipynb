```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
# matplotlib inline
```

```python
import json
with open('Thai_Cook.json') as outfile:
    Thai = json.load(outfile)
```

```python
text = ''
for i in range(1,len(Thai)+1):
    key = str(i)
    if len(Thai[key]['review']) > 0:
        text += Thai[key]['review']
```

```python
# Create and generate a word cloud image with default option:
wordcloud = WordCloud().generate(text)
```

```python
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# change some optional arguments of the WordCloud like max_font_size, max_word, and background_color
# lower max_font_size, change the maximum number of word and lighten the background:
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# create stopword list
stopwords = set(STOPWORDS)
stopwords.update(["food", "really", "always", "dishes", "little",'place'])
# Generate a word cloud image
wordcloud = WordCloud(max_words=50,stopwords=stopwords, background_color="white").generate(text)
# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Save the image in the img folder:
wordcloud.to_file("review.png")
```

```python
import json
with open('PLANET.json') as outfile:
    PLANET = json.load(outfile)
```

```python
PLANET['1']
```

```python
text = ''
for i in range(1,len(PLANET)+1):
    key = str(i)
    if PLANET[key]['rate'][6] == '1' or PLANET[key]['rate'][6] == '2':
        text += PLANET[key]['review']
```

```python
# select nouns from the text
import nltk
# function to test if something is a noun
is_noun = lambda pos: pos[:2] == 'NN'
tokenized = nltk.word_tokenize(text)
nouns = [word for (word, pos) in nltk.pos_tag(tokenized) if is_noun(pos)] 
```

```python
text = ''
for i in range(len(nouns)):
    text = text + ' ' + nouns[i]
```

```python
# create stopword list
stopwords = set(STOPWORDS)
stopwords.update(["food", "location", "service", "Translated", "food",'place','Original','Google','de','da'])
# Generate a word cloud image
wordcloud = WordCloud(max_words=50,stopwords=stopwords, background_color="white").generate(text)
# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
